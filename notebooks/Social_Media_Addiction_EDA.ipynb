```python
columns_to_drop = ['Student_ID']
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])
print(df.columns.tolist())
```

```python
# Group by age and calculate average daily usage
age_usage = df.groupby('Age')['Avg_Daily_Usage_Hours'].mean()
print(age_usage)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate: average age per category
age_avg = df.groupby('Affects_Academic_Performance')['Age'].mean().reset_index()

plt.figure(figsize=(6,5))
sns.barplot(data=age_avg, x='Affects_Academic_Performance', y='Age', color='skyblue')
plt.title('Average Age by Academic Performance Impact')
plt.xlabel('Affects Academic Performance')
plt.ylabel('Average Age')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Aggregate: average Addicted_Score and avg_daily_usage_hours per Affects_Academic_Performance
heatmap_data = df.groupby('Affects_Academic_Performance')[['Addicted_Score', 'Avg_Daily_Usage_Hours']].mean()

heatmap_data
```

```python
# Create categories for usage hours
bins = [0, 2, 4, 6, 8, 24]
labels = ['0-2h', '2-4h', '4-6h', '6-8h', '8+h']
df['usage_category'] = pd.cut(df['Avg_Daily_Usage_Hours'], bins=bins, labels=labels)

plt.figure(figsize=(8,5))
sns.countplot(x='usage_category', data=df, palette='Blues')
plt.title('Number of Students by Daily Usage Hours')
plt.xlabel('Daily Usage Category')
plt.ylabel('Number of Students')
plt.show()
```

```python
if 'Affects_Academic_Performance' in df.columns and 'Addicted_Score' in df.columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(x='Affects_Academic_Performance', y='Addicted_Score', data=df)
    plt.title('Addicted Score vs Affects Academic Performance')
    plt.xlabel('Affects Academic Performance')
    plt.ylabel('Addicted Score')
    plt.xticks(rotation=30)
    plt.show()
```

```python
if 'Addicted_Score' in df.columns and 'Mental_Health_Score' in df.columns:
    plt.figure(figsize=(7,6))
    sns.scatterplot(x='Addicted_Score', y='Mental_Health_Score', data=df)
    sns.regplot(x='Addicted_Score', y='Mental_Health_Score', data=df, scatter=False, ci=None)
    plt.title('Addicted Score vs Mental Health Score')
    plt.xlabel('Addicted Score')
    plt.ylabel('Mental Health Score')
    plt.show()

    corr = df['Addicted_Score'].corr(df['Mental_Health_Score'])
    print(f"\nPearson correlation between addicted_score and mental_health_score: {corr:.3f}")
```

```python
df_enc = df.copy()
cat_cols = df_enc.select_dtypes(include=['object','category']).columns.tolist()
if cat_cols:
    df_enc = pd.get_dummies(df_enc, columns=cat_cols, drop_first=True)

plt.figure(figsize=(12,10))
corr = df_enc.corr()
sns.heatmap(corr, cmap='coolwarm', center=0, linewidths=0.2)
plt.title('Correlation Matrix (numeric & encoded categorical)')
plt.show()
```
